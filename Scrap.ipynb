```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

url='https://pagalfree.com/category/Bollywood/'
url2='https://pagalfree.com/category/Indipop/'
# 0/2020.html

def Scrape(url,max_pages):
    
    current = 0
    year=1963
    
    count = 0
    while (year<2022):
        while( current<= max_pages) :
            new= str(current) +"/" +str(year)+'.html'
            print()
            current_url= url+str(new)
        
            print(current_url)
            html_text=requests.get(current_url).text
            Songs(html_text,year)
            count+=1
            print(count)
            current= current+1
        current=0
        year=year+1

def Songs(u,year):
    soup= BeautifulSoup(u, 'lxml')
    songs= soup.find_all('div', class_='col-lg-12 col-md-12 col-sm-12 col-xs-12 main_page_category_music_box')

    for song in songs:
        #Title
        title= song.find('b').text.replace(' ', ' ').strip()
        print(title)

        #Year
        yr= year

        #Type
        type='Bollywood'

        #Actors
        actors= song.find('p').text.replace(' ', ' ').strip()
        #print(singer)
        singer=[]
        #Singer
        if ',' in actors:
            count=1
        else:
            count=0

        if(count==1):
            singer=actors.split(',')
        #print(singer[0])
        else:
            singer.append("None")

        #Album

        album= singer[0]+actors
        #print(album)

        #Genre
        if singer[0].endswith('i'):
            genre='Instrumental'
        elif singer[0].endswith('b'):
            genre='jazz'
        elif singer[0].endswith('c'):
            genre='Rock'
        elif singer[0].endswith('d'):
            genre='Pop'
        elif singer[0].endswith('e'):
            genre='Soul'
        elif singer[0].endswith('f'):
            genre='Dance'
        elif singer[0].endswith('g'):
            genre='Ryhtem'
        elif singer[0].endswith('h'):
            genre='Blues'
        elif singer[0].endswith('a'):
            genre='JazzPop'

        elif singer[0].endswith('j'):
            genre='Soulistry'
        elif singer[0].endswith('k'):
            genre='Makeover'
        elif singer[0].endswith('l'):
            genre='Anthem'
        elif singer[0].endswith('m'):
            genre='Soul'
        elif singer[0].endswith('n'):
            genre='Country'
        else:
            genre='None'
        #print(genre)
        
        
        data = {'Song Title':[title],'Year':[yr],'Song Type':[type], 'Actors':[actors] , 'Singer':[singer[0]],'Album':[album],'Genre':[genre]}
        df = pd.DataFrame(data) 
        df.to_csv('Songs.csv', mode='a',index=False, header=False, encoding='utf-8')
    

Scrape(url2,5)  

#https://pagalfree.com/category/Bollywood/0/1963.html
 
```
