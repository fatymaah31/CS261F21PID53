```python
from selenium import webdriver
from bs4 import BeautifulSoup
from bs4 import *
import pandas as pd
from csv import writer
from time import sleep
import re
import csv
import collections
import sys
#install chrom webdriver
#webdriver can be downloaded from #https://chromedriver.storage.googleapis.com/index.html?path=94.0.4606.41/
#driver = webdriver.Chrome(executable_path='chromedriver.exe')

#arrays_of_all_entities
song_arr=[] 
singer=[]
genre = []
subgenre = []
price = []
tags=[]

#genre_hiphop
def soundclick_hiphop():
    sub_ID = [ 95, 154, 91, 98, 94, 99  ]
    pages = [  385, 109, 3980, 564, 163, 246]
    sub_genre = [ "Free Style", "Hip-Hop Asian" , "Hip-Hop General" , "New School", "Christian Rap" , "Positive Vibes" ]
    
    for id in range(len(sub_ID)):
        print(sub_genre[id])
        for page in range(1,int(pages[id]),6):
            print(page)
            driver = webdriver.Chrome(executable_path='C:\chromedriver.exe')
            #print(" opening Browser")
            driver.get('https://www.soundclick.com/music/charts.cfm?genreID=19&subgenreID='+str(sub_ID[id])+'&currentPage='+str(page))
            #sleep(2)
            content = driver.page_source
            soup = BeautifulSoup(content)
            #print(soup)
            for a in soup.findAll('div',attrs={'class':'charts_songBox'}):
                #songs = soup.find('div',attrs={'class':'charts_songname songActivityIndicator clk_loadPlaylist'})
                #songs = soup.select_one("[id='charts_titleBox'] > span:nth-of-type(1000)")
                #songs=re.sub(' \n ',' ',(soup.find('span',attrs={'class':'charts_titleBox'}).getText())
                ##numbers = a.find('div', attrs={'class':'chartsPosition'})  
                
                    #singer
                try:

                    singers = a.find('a', attrs={'class':'sclk_link'})
                    singer.append(singers.text)
                except:
                    singer.append("No singer Info")
                     
                
                    #song
                song=a.find('div', class_='charts_titleBox')
                s=song.find('span').text
                song_arr.append(s)
                
                    #genre
                genre.append("Hip Hop")
                
            
                   #price
                try:
                    
                    ok = a.find('div', class_='charts_purchase clk_openFancy')
                    je= ok.find('span').text
                    price.append(je)
                except:
                    price.append("  Free")
                    
                  #tags
                try:
                    
                    k = a.find('div', class_='charts_tags tonedDown')
                    l=k.find('span').text
                    if(l==""):
                        tags.append("No tags")
                    else:
                        tags.append(l)
                except:
                    tags.append("No Tags")
                
                subgenre.append(sub_genre[id])

#genre_instrumental
def soundclick_beats():
    
    sub_ID=["262","186","249","168","264","169","167","188","47"]
    pages=["338","134","212","2424","125","2300","200","220","168"]
    sub_genre= ["Club Bangas", "Jazzy Beats", "Instrumental with hooks", "HipHop", "Hard Core" , "Beats General", "Electro-hop", 
                "Pop", "Film Music" ]
    for id in range(len(sub_ID)):
        print(sub_genre[id])
        for page in range(1,int(pages[id]),6):
            print(page)
            driver = webdriver.Chrome(executable_path='C:\chromedriver.exe')
            #print(" opening Browser")
            driver.get('https://www.soundclick.com/music/charts.cfm?genreID=7&subgenreID='+str(sub_ID[id])+'&currentPage='+str(page))
            #sleep(2)
            content = driver.page_source
            soup = BeautifulSoup(content)
            #print(soup)
            for a in soup.findAll('div',attrs={'class':'charts_songBox'}):
                #songs = soup.find('div',attrs={'class':'charts_songname songActivityIndicator clk_loadPlaylist'})
                #songs = soup.select_one("[id='charts_titleBox'] > span:nth-of-type(1000)")
                #songs=re.sub(' \n ',' ',(soup.find('span',attrs={'class':'charts_titleBox'}).getText())
                ##numbers = a.find('div', attrs={'class':'chartsPosition'})  
                
                    #singer
                try:

                    singers = a.find('a', attrs={'class':'sclk_link'})
                    singer.append(singers.text)
                except:
                    singer.append("No singer Info")
                     
                
                    #song
                song=a.find('div', class_='charts_titleBox')
                s=song.find('span').text
                song_arr.append(s)
                
                    #genre
                genre.append("Beats and instruments")
                
               # bpm.append(100)
               # number.append(numbers.text)
            
                   #price
                try:
                    
                    ok = a.find('div', class_='charts_purchase clk_openFancy')
                    je= ok.find('span').text
                    price.append(je)
                except:
                    price.append("  Free")
                    
                  #tags
                try:
                    
                    k = a.find('div', class_='charts_tags tonedDown')
                    l=k.find('span').text
                    if(l==""):
                        tags.append("No tags")
                    else:
                        tags.append(l)
                except:
                    tags.append("No Tags")
                
                subgenre.append(sub_genre[id])
            
                

soundclick_hiphop()
soundclick_beats()
df = pd.DataFrame({'Songs Title': song_arr , 'Genre': genre,'Sub Genre': subgenre  , 'Artist/Album Name' : singer , 'Album Price': price , 'Tags':tags})
df.to_csv('songs2.csv',mode='a' , index=False, encoding='utf-8')
print("Check the csv file")
```
