```python
from selenium import webdriver
from bs4 import BeautifulSoup
from bs4 import *
import pandas as pd
from csv import writer
from time import sleep

#install chrom webdriver
#webdriver can be downloaded from #https://chromedriver.storage.googleapis.com/index.html?path=94.0.4606.41/
#driver = webdriver.Chrome(executable_path='chromedriver.exe')


def beatstar():
    song=[] #List to store name of the product
    bpm=[] #List to store price of the product
    singer=[]
    empty = []
    genre=["POP", "HIP_HOP" ,"TRAP" ,"RNB" , "OLD_SCHOOL", "UNDERGROUND", "WEST_COAST"]
    for index in range(len(genre)):
        driver = webdriver.Chrome(executable_path='chromedriver.exe')
        sleep(2)
        driver.get("https://www.beatstars.com/search?type=tracks&genres="+genre[index])
        print("Browser closed")
        content = driver.page_source
        soup = BeautifulSoup(content)
        #sleep(randint(2,10))
        # print(soup)
        # list-item ng-star-inserted
        # item-title
        # name ng-star-inserted
        # name ng-star-inserted
        print("appending songs for" + genre[index])
        for a in soup.findAll('mp-card-figcaption-template',attrs={'class':'list-item'}):
            songs=a.find('a', attrs={'class':'ng-star-inserted'})
            beat=soup.find('span', attrs={'class':'name ng-star-inserted'})
            singers=soup.find('div', attrs={'class':'wrapper flex ng-star-inserted'})

            song.append(songs.text)
            bpm.append(beat.text)
            singer.append(singers.text)
        print (song)
        print("adding data to files")
        #ratings.append(rating.text)
        df = pd.DataFrame({'Songs Name':song , 'Genre': genre[index] , 'Beat per Minute': bpm , 'Singer' : singer })
        df.to_csv('songs.csv',mode='a' , index=False, encoding='utf-8')

def soundclick_hiphop():
    song=[] #List to store name of the product
    bpm=[] #List to store price of the product
    singer=[]
    genre = []

    for page in range(3):
        driver = webdriver.Chrome(executable_path='chromedriver.exe')
        print("Browser opened")
        driver.get("https://www.soundclick.com/music/charts.cfm?genreID=19&currentPage="+str(page))
        print("Browser closed")  
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'charts_titleBox'}):
            #songs = soup.find('div',attrs={'class':'charts_songname songActivityIndicator clk_loadPlaylist'})
            sleep(2)
            songs = soup.select_one("[id='charts_titleBox'] > span:nth-of-type(1000)")
            singers = a.find('a', attrs={'class':'sclk_link'})
            print(a.get_text())
            song.append(a.get_text())
            singer.append(singers.text)
            genre.append("Hip Hop")
            bpm.append(100)
            #print(song)
            
            
            df = pd.DataFrame({'Songs Name':song , 'Genre': genre ,  'Beat per Minute': 100 , 'Singer' : singer })
            df.to_csv('songs.csv',mode='a' , index=False, encoding='utf-8')
#             with open('songs.csv', 'a') as f_object:
#                 writer_object = writer(f_object)
#                 writer_object.writerow(song)
#                 f_object.close()

def soundclick_acoustic():
    song_arr=[] #List to store name of the product
    bpm=[] #List to store price of the product
    singer=[]
    genre = []

    for page in range(4):
        driver = webdriver.Chrome(executable_path='chromedriver.exe')
        print("Scraping "+str(page+1)+" pages out of 1000000s")
        driver.get('https://www.soundclick.com/music/charts.cfm?genreID=1&currentPage='+str(page))
        sleep(2)
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div',attrs={'class':'charts_titleBox'}):
            #songs = soup.find('div',attrs={'class':'charts_songname songActivityIndicator clk_loadPlaylist'})
            songs = soup.select_one("[id='charts_titleBox'] > span:nth-of-type(1000)")
            singers = a.find('a', attrs={'class':'sclk_link'})
            s=a.get_text()
            song_arr.append(s)
            print (song_arr)
            singer.append(singers.text)
            genre.append("Acoustic")
            bpm.append(100)
            df = pd.DataFrame({'Songs Name': song_arr , 'Genre': genre ,  'Beat per Minute': 100 , 'Singer' : singer })
            df.to_csv('songs.csv',mode='a' , index=False, encoding='utf-8')
#beatstar()
#soundclick_hiphop()
soundclick_acoustic()
print ("Check the csv file")
```
